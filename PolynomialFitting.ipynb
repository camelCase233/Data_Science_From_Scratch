```python
def get_column_mat(degree_req: int, inputs:List, degree: int)->List:
    x_val, y_val = zip(*inputs)
    col_mat = []
    for i in range(degree+1):
        sum = 0
        for j in range(len(x_val)):
            sum += (x_val[j][0]**i) * (x_val[j][0]**degree_req)
        col_mat.append(sum)
    return col_mat
```

```python
def get_w(inputs:List, degree:int):
    A = np.array([
        get_column_mat(i, inputs, degree)
        for i in range(degree+1)
    ])
    return A
```

```python
def get_sol_mat(inputs: List, degree: int)->List:
    x_val,y_val = zip(*inputs)
    output = []
    for i in range(degree+1):
        sum = 0
        for j in range(len(y_val)):
            sum += (y_val[j][0] * (x_val[j][0]**i))
        output.append(sum)
    return np.array(output).reshape(degree+1, 1)
```

```python
def get_sol_w(inputs: List, degree: int):
    return np.linalg.solve(get_w(inputs, degree), get_sol_mat(inputs, degree))
```

```python

m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)

inputs = list(zip(X, y))

x_test_values = np.linspace(-3, 3)  # Smooth range
degree = 2
weights = get_sol_w(inputs, degree)

# Build design matrix
X_matrix = np.array([[x**i for i in range(degree + 1)] for x in x_test_values])
y_predicted = X_matrix @ weights
y_predicted = y_predicted.flatten()

plt.scatter(X, y, label="Data")
plt.plot(x_test_values, y_predicted, color='r', label=f"Degree {degree} Fit")
plt.legend()
plt.title("Polynomial Regression Fit (Degree 1)")
plt.show()
```
